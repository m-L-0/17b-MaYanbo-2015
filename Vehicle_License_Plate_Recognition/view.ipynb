```python
import matplotlib.pyplot as plt
import os
import numpy as np

def do_the_statistics(path_dir):
    dir_lst = os.listdir(path_dir)
    #print(dir_lst)
    counter = {}
    for i in dir_lst:
        counter[i] = len(os.listdir(path_dir+i))
        #print (counter[i])
    print(counter)
    N = 34
    ind = np.arange(N)
    width = 0.35
    fig, ax = plt.subplots()
    ax.bar(ind, counter.values())
    ax.set_xlabel('labels')
    ax.set_ylabel('quantity')
    ax.set_title(' data distribution ')
    ax.set_xticks(ind + width)
    ax.set_xticklabels(list(counter.keys()))
    fig.tight_layout()
    plt.show()
```

```python
do_the_statistics('data/')
```

```python
def do_the_pie(path_dir):
    dir_lst = os.listdir(path_dir)
    #print(dir_lst)
    counter = {}
    for i in dir_lst:
        counter[i] = len(os.listdir(path_dir+i))
        #print (counter[i])
    print(counter)
    N = 34
    ind = np.arange(N)
    width = 0.35
    fig, ax = plt.subplots()
    rects=ax.pie(counter.values(), autopct='%1.1f%%', shadow=True,labels=counter.keys(), labeldistance=1.1, startangle = 90,pctdistance = 0.6)

    ax.set_title(' data distribution ')
    fig.tight_layout()
    plt.show()
```

```python
do_the_pie('data/')
```
